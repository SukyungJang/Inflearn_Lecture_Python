### ê³¼ì œ
- ì£¼ì‹ í˜„ì¬ê°€ í¬ë¡¤ë§í–ˆë˜ ë°ì´í„°ë¥¼
- ì•„ë˜ ì–‘ì‹ì˜ ì—‘ì…€ì„ ë¶ˆëŸ¬ì™€ì„œ B2, B3, B4ì— ì €ì¥

```python
import requests
from bs4 import BeautifulSoup
import openpyxl

fpath = r'C:\Users\82104\OneDrive\ë°”íƒ• í™”ë©´\jangcoding\inflearn\ì´ê²ƒì´ í¬ë¡¤ë§ì´ë‹¤ - ê¸°ë³¸í¸\data.xlsx'
wb = openpyxl.load_workbook(fpath)
ws = wb.active # í˜„ì¬ í™œì„±í™”ëœ ì‹œíŠ¸ ì„ íƒ

# ì¢…ëª© ì½”ë“œ ë¦¬ìŠ¤íŠ¸
codes = [
    '005930',
    '000660',
    '035720'
]
row = 2
for code in codes:
    url = f"https://finance.naver.com/item/sise.naver?code={code}"
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    price = soup.select_one('#_nowVal').text
    price = price.replace(',', '')
    print(price)
    ws[f'B{row}'] = int(price)
    row = row + 1

wb.save(fpath)
```
